```python
from google.colab import files

uploaded = files.upload()
pdf_file = list(uploaded.keys())[0]
print("PDF carregado:", pdf_file)
```

```python
import os
from pdf2image import convert_from_path, pdfinfo_from_path
from tqdm import tqdm

# Detectar automaticamente o número de páginas
info = pdfinfo_from_path(pdf_file)
total_paginas = info["Pages"]

print("Total de páginas detectadas:", total_paginas)

os.makedirs("pages", exist_ok=True)
image_paths = []

for i in tqdm(range(1, total_paginas + 1), desc="Convertendo páginas"):
    try:
        page = convert_from_path(pdf_file, dpi=200, first_page=i, last_page=i)[0]
        img_path = f"pages/page_{i}.png"
        page.save(img_path, "PNG")
        image_paths.append(img_path)
    except Exception as e:
        print(f"Erro na página {i}: {e}")

print("Total de páginas convertidas:", len(image_paths))
```

```python
import easyocr

reader = easyocr.Reader(['en'], gpu=True)
print("EasyOCR carregado!")
```

```python
from deep_translator import GoogleTranslator

translator = GoogleTranslator(source='en', target='pt')

resultados = []

for page_num, img in enumerate(image_paths, start=1):
    print(f"OCR na página {page_num}...")
    ocr_result = reader.readtext(img, detail=1)

    caixas = []
    for box, text, conf in ocr_result:
        x_coords = [p[0] for p in box]
        y_coords = [p[1] for p in box]
        x_min, y_min = min(x_coords), min(y_coords)

        caixas.append({
            "texto": text,
            "conf": conf,
            "x": x_min,
            "y": y_min
        })

    caixas_ordenadas = sorted(caixas, key=lambda c: (c["y"], c["x"]))

    resultados.append({
        "pagina": page_num,
        "caixas": caixas_ordenadas
    })

print("OCR concluído!")
```

```python
with open("manga_traduzido.txt", "w", encoding="utf-8") as f:
    for pagina in resultados:
        f.write(f"\n\n=== Página {pagina['pagina']} ===\n")

        for caixa in pagina["caixas"]:
            original = caixa["texto"]
            traduzido = translator.translate(original)

            f.write(f"\n[EN] {original}")
            f.write(f"\n[PT] {traduzido}\n")

print("Arquivo gerado: manga_traduzido.txt")
```
